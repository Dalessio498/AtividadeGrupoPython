```python
# --- Passo 2: Mesclar duas tabelas e criar uma coluna de valor total ---

# Define o ladode cada tabela e o lado esquerdo como principal
df_medio = pd.merge(right=df_clientes, left=df_vendas, on="ID_Cliente", how="left")
df_completo = pd.merge(df_medio, df_produtos, on="ID_Produto")

# Define o valor da coluna 'valorTotal' como o preço * quantidade ou vice-versa
df_completo['valorTotal'] = df_completo['Quantidade'] * df_completo['Preco_Unitario']

display (df_completo)

#-------------------------------------------------------------------------
```

```python
# --- Passo 3: Criar uma função que vá determinar a prioridade de entrega por meio do valor

# Cria a função classicarVendas
def classificarVendas(valor):

    # Condiciona que se o valor for maior que 500 R$, a entrega ser do tipo 'Expresso'

    if valor > 500:
        return 'Expresso'
    else:
        return 'Normal'
    

# Aplica a função (classificarVendas) na tabela 'Prioridade' com base no resultado
df_completo['Prioridade'] = df_completo['Preco_Unitario'].apply(classificarVendas)
display(df_completo)


#------------------------------------------------------------------------------------------
```

```python
# --- Passo 4: Exercicios de mostragem de dados ---

# --- 1. Qual o faturamento total (soma do Valor_Total) por Categoria de produto? ---

faturamento_Total = df_completo.groupby('Categoria')['valorTotal'].sum()
print("\n" + "="*80 + "\n")
print(" Exercicio 1 - Faturamento total por categoria")
print("\n" + "="*80 + "\n")
display(faturamento_Total.reset_index())

# --- 2. Qual a quantidade total de produtos vendidos por Estado do cliente? ---

qtd_produtos_uf = df_completo.groupby('Estado')['Quantidade'].sum()
print("\n" + "="*80 + "\n")
print(" Exercicio 2 - Quantidade total de produtos vendidos por estado")
print("\n" + "="*80 + "\n") 
display (qtd_produtos_uf.reset_index())

# --- 3. Qual o faturamento total por Cliente? (Listar o top 3) ---

faturamentoTotalCliente = df_completo.groupby('Nome_Cliente')['valorTotal'].sum()
top3 = faturamentoTotalCliente.head(3).sort_values(ascending=False) # Deixa em ordem decrescente
print("\n" + "="*80 + "\n")
print(" Exercicio 3 - Faturamento total por cliente")
print("\n" + "="*80 + "\n")
display (top3.reset_index())


# --- 4. (Bônus) Qual o produto mais vendido (em Quantidade) da categoria 'Eletrônico'? (Exigirá filtro +groupby) ---

produtoMaisVendido = df_completo.groupby('Quantidade')['Categoria'].count()

filtro_eletronico = df_completo[df_completo["Categoria"] == "Eletrônico"]
mais_vendidos_eletronico = (filtro_eletronico.groupby("Nome_Produto")["Quantidade"].sum().sort_values(ascending=False))
produto_top = mais_vendidos_eletronico.head(1)

print("\n" + "="*80 + "\n")
print(" Exercicio 4 - Produto mais vendido em quantidade da categoria eletronico")
print("\n" + "="*80 + "\n")
display(produto_top.reset_index())
print("\n" + "="*80 + "\n")

#-----------------------------------------------------------------------------------------------------
```

```python
# --- Passo 5: Criar tabela de barras do df_completo de vendas

plt.figure(figsize=(10, 6))

paletaCores = ["#5c65c0", "#6f95ff", "#413b6b"]

sns.set_theme(style="darkgrid")

sns.barplot(
    data=df_completo,
    x='Categoria',
    y='valorTotal',
    palette=paletaCores,
    estimator=sum,
    errorbar=None
)

plt.title('Faturamento Por Categoria', fontsize=16, fontweight='bold'),
plt.xlabel('Categoria', fontsize=12),
plt.ylabel('Valor Total', fontsize=12)
plt.show()

#------------------------------------------------------------------
```
